```python
#2(#import PIL로 불러올 시,Image 모듈은 안불러와져서 에러가 뜸)
from PIL import Image 
```

```python
#4 이미지 크기 변경
img_resized = img.resize((700,700), resample=0)
#크기 변경된 이미지 출력
img_resized.show()
#원본 이미지 출력
img.show()
```

```python
#5. 이미지 자르기
img_cropped = img.crop((100,100,500,500))
img_cropped.show()
```

```python
#6. 이미지 회전
img_rotated = img.rotate(180)
img_rotated.show()
```

```python
#7. 이미지 대칭
# 좌우대칭
img_flipped_LR = img.transpose(Image.FLIP_LEFT_RIGHT)
img_flipped_LR.show()

# 상하대칭
img_flipped_TB = img.transpose(Image.FLIP_TOP_BOTTOM)
img_flipped_TB.show()
```

```python
#8. 이미지에 필터 부여하기
from PIL import ImageFilter
'''
blur,contour,detail,smooth등등 제공
'''
```

```python
#8-1. 블러효과
# Blur 효과를 줄 때 가장 많이 사용하는 가우시안 블러 --> 숫자를 크게 할 수록 흐려짐
img_blur = img.filter(ImageFilter.GaussianBlur(10))
# 이미지 출력
img_blur.show()
```

```python
#8-2. 컨투어 효과
'''
컨투어(contour)란 동일한 색 또는 동일한 픽셀값(강도,intensity)을 가지고 있는 영역의 경계선 정보다. 물체의 윤곽선, 외형을 파악하는데 사용된다.
'''
# 컨투어 필터 정의
contour_filter = ImageFilter.Kernel((3, 3), (-1, -1, -1, -1, 8, -1, -1, -1, -1), scale=1)
# 컨투어 효과 적용
img_contour = img.filter(contour_filter)
# 이미지 출력
img_contour.show()
```

```python
#8-3. 디테일 효과
'''
디테일 효과는 이미지의 세부 사항을 강조하는 효과를 적용합니다.
이미지의 미세한 세부 사항 강조, 이미지의 고주파 성분을 강조하여 뾰족한 특징을 부각합니다.
'''
# 디테일 효과 적용
img_detail = img.filter(ImageFilter.DETAIL)
img_detail.show()
```

```python
#8-4. 스무스 효과
'''
스무스 효과는 이미지의 부드러운 효과를 적용합니다.
이미지의 노이즈를 제거하고 부드러운 외형을 만듭니다.
'''
# 스무스 효과 적용
img_smooth = img.filter(ImageFilter.SMOOTH)
img_smooth.show()

# SMOOTH_MORE 효과 적용
'''
스무스 효과보다 좀 더 강력하나 과도하게 들어가면 이미지를 흐리게 만들 수 있습니다.
'''
img_smooth = img.filter(ImageFilter.SMOOTH_MORE)
img_smooth.show()
```

```python
#9. 이미지 합치기
from PIL import Image
 
image1 = Image.open('homer.jpg')
image1.show()
 
image2 = Image.open('Lisa.jpg')
image2.show()
image1 = image1.resize((700,700))
image2 = image2.resize((700,700))
image1_size = image1.size
image2_size = image2.size
new_image = Image.new('RGB',(2*image1_size[0], image1_size[1]), (250,250,250))
new_image.paste(image1,(0,0))
new_image.paste(image2,(image1_size[0],0))

#10. 이미지 저장
new_image.save("homer and Lisa.jpg","JPEG")
new_image.show()
```

```python
#11. 이미지 바이트 배열로 변환하기
import cv2

# 이미지 파일을 읽어들입니다.
image = cv2.imread('homer and Lisa.jpg')

# 이미지를 바이트로 변환합니다.
succ, enc_image = cv2.imencode('.jpg', image)
image_bytes = enc_image.tobytes()

# 변환된 이미지 바이트를 출력합니다.
print(image_bytes)
a = image_bytes

#12. 바이트 배열로 이미지 읽기
import numpy as np
# 이미지 바이트를 읽습니다.
image_bytes = a  # 이미지 바이트 배열을 여기에 넣으십시오.

# 바이트를 NumPy 배열로 변환합니다.
image_array = np.frombuffer(image_bytes, dtype=np.uint8)

# NumPy 배열을 이미지로 변환합니다.
decoded_image = cv2.imdecode(image_array, cv2.IMREAD_COLOR)

# 이미지를 표시하거나 저장하거나 다른 작업을 수행합니다.
cv2.imshow("Decoded Image", decoded_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# 이미지를 파일로 저장하려면
# cv2.imwrite("output_image.jpg", decoded_image)
```
